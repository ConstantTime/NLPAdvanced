```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('lena.png')
# image used in all CV courses
```

```python
plt.imshow(img)
plt.show()
```

```python
# our task is to make it black and white
bw = img.mean(axis = 2)
plt.imshow(bw , cmap = 'gray')
plt.show()
```

```python
# creating a gaussian filter
N = 20
center = 9.5
W = np.zeros((N , N))
for i in range(N) :
    for j in range(N) :
        dist = (i - center) ** 2 + (j - center) ** 2
        W[i , j] = np.exp(-dist / 50)
        
W /= W.sum() # normalize the kernel
plt.imshow(W , cmap = 'gray')
plt.show()
```

```python
# output will store the blurred image
output = convolve2d(bw , W)
plt.imshow(output , cmap = 'gray')
plt.show()
```

```python
print(output.shape)
out3 = np.zeros(img.shape)
print(out3.shape)
```

```python
for i in range(3):
    out3[:,:,i] = convolve2d(img[:,:,i], W, mode='same')
plt.imshow(out3)
plt.show()
```
